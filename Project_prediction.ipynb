```python
#read the data from the csv file with pandas library
%pip install pandas 
import pandas as pd
df=pd.read_csv("prices.csv",)
df #display the dataframe of pandas
```

```python
#get infos from about the not_null columns of the dataframe
df.info()
```

```python
#do aggregate to know how many null values in the dataframe df
df.isnull().sum()
```

```python
#drop all the null columns in the dataframe 
df.dropna(inplace=True)
```

```python
#just a verification 
df.isnull().sum()
```

```python
#replace the "," and transform the New price ($) Column,and convert it to float
df["New Price ($)"]=df["New Price ($)"].str.replace(",","").astype(float)
```

```python
#replace the "," and transform the Used price ($) Column,and convert it to float
df["Used Price ($)"]=df["Used Price ($)"].str.replace(",","").astype(float)
```

```python
#replace the "," and transform the Units Sold Column,and convert it to float
df["Units Sold"]=df["Units Sold"].str.replace(",","").astype(float)
```

```python
#strip the "%" and convert the percentage to float in Inflation Rate (%) Column
df["Inflation Rate (%)"]=df["Inflation Rate (%)"].str.rstrip("%").astype(float)/100
```

```python
#strip the "%" and convert the percentage to float  in  and the Interest Rate (%) Column
df["Interest Rate (%)"]=df["Interest Rate (%)"].str.rstrip("%").astype(float)/100
```

```python
#fix the format of the Date Column 
df["Month/Year"]=pd.to_datetime(df["Month/Year"],format="%y-%b")
```

```python
#do a visualization to know the difference between the New and Used Price Over Time
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.plot(df["Month/Year"],df["New Price ($)"],label="New Price")
plt.plot(df["Month/Year"],df["Used Price ($)"],label="Used Price")
plt.title("Price Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Price($)")
plt.legend()
plt.savefig("plot.png")
plt.show()
```

```python
#calculate the correlation matrix to retrieve some information about the relation between the factors in the dataset
df.corr()
```

```python
#visualization 
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
sns.heatmap(df.select_dtypes(include=['float64','int64']).corr(),annot=True,cmap="coolwarm")
plt.title("Correlation Matrix")
plt.savefig("heatMap.png")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
X=df[["New Price ($)","Used Price ($)"]]
Y=df["Units Sold"]
model=LinearRegression()
model.fit(X,Y)
prediction=model.predict(X)
print("Prediction for Units Sold:\n",prediction)
from sklearn.metrics import r2_score
score=r2_score(Y,prediction)
print("the r2 score:",score)
from sklearn.metrics import mean_absolute_error
mean_score=mean_absolute_error(Y,prediction)
print("mean_score:",mean_score)
from sklearn.metrics import mean_squared_error
import numpy as np
mse=mean_squared_error(Y,prediction)
rmse=np.sqrt(mse)
print("mean squared error:",mse)
print("root mean squared error:",rmse)
```

```python
import matplotlib.pyplot as plt
df["Predicted Units Sold"] = prediction
new_df=df.to_csv("new_data.csv")
plt.plot(df.index, df["Units Sold"], label="Actual Units Sold", marker="o")
plt.plot(df.index, df["Predicted Units Sold"], label="Predicted Units Sold", linestyle="--", marker="o")
plt.xlabel("Time Steps")
plt.ylabel("Units Sold")
plt.title("Actual vs Predicted Units Sold Over Time")
plt.savefig("Predicted.png")
plt.legend()
plt.show()
```
